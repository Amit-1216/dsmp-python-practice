```python
# find ranking of each movie in the group according to IMDB score
def rank_movie(group):
  group['genre_rank'] = group['IMDB_Rating'].rank(ascending=False)
  return group
```

```python
ranked = genres.apply(rank_movie)
drama_movies = ranked[ranked['Genre'] == 'Drama']
drama_movies.sort_values('genre_rank')
```

```python
# find normalized IMDB rating group wise
def normalized_rating (group):
  group['normalized_rating'] = ((group['IMDB_Rating'] - group['IMDB_Rating'].min()) / (group['IMDB_Rating'].max() - group['IMDB_Rating'].min()))
  return group

genres.apply(normalized_rating)
```

```python
# agg on multiple groupby
duo.agg(['min','max','mean'])
```

```python
ipl = pd.read_csv('/content/deliveries.csv')
ipl.sample(5)
```

```python
ipl.columns
```

```python
# find the top 10 batsman in terms of runs
ipl.groupby('batsman')['batsman_runs'].sum0().sort_values(ascending=False).head(10)
```

```python
# find the batsman with max no of sixes
six = ipl[ipl['batsman_runs'] == 6]
six.groupby('batsman')['batsman_runs'].size().sort_values(ascending=False).head(1)

## ----------------------------------------------------------------------------------------

# ipl[ipl['batsman_runs'] == 6]['batsman'].value_counts().head(1)
```

```python
# find batsman with most number of 4's and 6's in last 5 overs
# over = ipl[(ipl['over'] > 15) & ((ipl['batsman_runs'] == 4) | (ipl['batsman_runs'] == 6))]
# over['batsman'].value_counts().head(1)

## -------------------------------------------------------------------------------------------------

ipl[(ipl['over'] > 15) & ((ipl['batsman_runs'] == 4) | (ipl['batsman_runs'] == 6))].groupby('batsman')['batsman_runs'].size().sort_values(ascending=False).head(1)
```

```python
# find V Kohli's record against all teams
v = ipl[ipl['batsman'] == 'V Kohli']
v.groupby('bowling_team')['batsman_runs'].sum().reset_index()
```

```python
# Create a function that can return the highest score of any batsman
def highest_score(player):
    return (
        ipl[ipl['batsman'] == player]
        .groupby('match_id')['batsman_runs']
        .sum()
        .max()
    )


highest_score('MS Dhoni')
```
