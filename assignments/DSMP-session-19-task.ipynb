
###`Q-8:` Make a dataframe for all batting pairs played together.
```
Batsman1 Batsman2 Runs Avg StrikeRate
```

> Just to ease this question you can count wide-balls for strike rate.

````python
temp = pd.DataFrame([])
ipl_df["partnership"] = ipl_df.apply(lambda x : tuple(sorted([x["batter"], x["non-striker"]])), axis=1)
batting_pairs = ipl_df.groupby("partnership")
batting_pairs
````

````python
temp["Runs"] = batting_pairs["total_run"].sum()
temp["Balls"] = batting_pairs.size()
temp["Avg"] = batting_pairs["total_run"].mean()
temp["StrikeRate"] = (batting_pairs["total_run"].sum() / batting_pairs.size()) * 100

temp = temp.reset_index()
temp["Batsman1"] = temp["partnership"].str[0]
temp["Batsman2"] = temp["partnership"].str[1]

temp = temp[["Batsman1", "Batsman2", "Runs", "Avg", "StrikeRate"]]

temp.head()
````

````python
temp = pd.DataFrame([])

ipl_df["partnership"] = ipl_df.apply(
    lambda x: tuple(sorted([x["batter"], x["non-striker"]])),
    axis=1
)

batting_pairs = ipl_df.groupby("partnership")

# Total runs scored together
temp["Runs"] = batting_pairs["total_run"].sum()

# Number of balls together
temp["Balls"] = batting_pairs.size()

# Average runs per ball together
temp["Avg"] = batting_pairs["total_run"].mean()

# Strike Rate
temp["StrikeRate"] = (temp["Runs"] / temp["Balls"]) * 100

temp = temp.reset_index()

temp["Batsman1"] = temp["partnership"].str[0]
temp["Batsman2"] = temp["partnership"].str[1]

temp = temp[["Batsman1", "Batsman2", "Runs", "Avg", "StrikeRate"]]

temp.head()
````
